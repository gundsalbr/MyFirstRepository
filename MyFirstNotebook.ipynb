# Exploración de análisis de datos en Python

## 1. Importando librerías requeridas 

```python
import pandas as pd
import numpy as np
import seaborn as sns #visualización
import matplotlib.pyplot as plt #visualización
%matplotlib inline
sns.set(color_codes=True)
```

## 2. Cargando datos

```python
df = pd.read_csv("data.csv")
#Despliega primeras 5 filas
df.head(6)
```

```python
#despliega últimas n filas, 15 en este caso
df.tail(15)
```

## 3. Revisando tipos de datos

```python
df.dtypes
```

## 4. Omitir columnas irrelevantes

```python
df=df.drop(['Engine Fuel Type', 'Market Category', 'Vehicle Style', 'Popularity', 'Number of Doors', 'Vehicle Size'],axis=1)
df.head(5)
```

## 5. Renombrar columnas

```python
df = df.rename(columns={"Engine HP": "HP", "Engine Cylinders":"Cylinders",  "Transmission Type": "Transmission", "Driven_Wheels": "Drive Mode"})
df.head(5)
```

## 6. Remover filas duplicadas

```python
df.shape #muestra el número de filas y columnas
```

```python
duplicate_rows_df = df[df.duplicated()]
print("número de filas duplicadas: ", duplicate_rows_df.shape)
```

```python
#Contar datos duplicados
df.count()
```

```python
df=df.drop_duplicates()
df.head(5)
```

```python
df.count()
```

##  7. Descartar Valores faltantes o nulos

```python
print(df.isnull().sum()) #Cuenta el número de datos faltantes o nulos
```

```python
df =df.dropna() #descarta valores faltantes 
df.count()
```

```python
print(df.isnull().sum()) #Se verifica que no existan valores faltantes o nulos
```
